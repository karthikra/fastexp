```python
bears = DataBlock(blocks=(ImageBlock,CategoryBlock),
                  get_items=get_image_files,
                  splitter=RandomSplitter(valid_pct=0.3,seed=42),
                 get_y=parent_label,
                 item_tfms=Resize(128))
```

```python
dls = bears.dataloaders(path_image)
```

```python
dls.valid.show_batch(max_n=4,rows=1)
```

```python
bears = bears.new(item_tfms=RandomResizedCrop(128,min_scale=0.3))
```

```python
dls = bears.dataloaders(path_image)
dls.train.get_idxs = lambda: Inf.ones
dls.train.show_batch(max_n=4, rows=1)
```

```python
dls.valid.show_batch(max_n=4,rows=1)
```

###### Using Data Augmentation

```python
bears = bears.new(item_tfms=Resize(128),batch_tfms=aug_transforms(mult=2.0))
dls = bears.dataloaders(path_image)
dls.train.get_idxs = lambda: Inf.ones
dls.train.show_batch(max_n=8, rows=2)
```

###### Training the MOdel and Using the result to clean the Data

```python
bears = bears.new(item_tfms=RandomResizedCrop(224,min_scale=0.5),batch_tfms=aug_transforms())
dls = bears.dataloaders(path_image)
```

```python
learn = cnn_learner(dls,arch=resnet34,metrics=error_rate)
```

```python
# learn.fine_tune(10)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

##### Creating a Notebook app from the model

```python
btn_upload = widgets.FileUpload()
btn_upload
```

```python
img = PILImage.create(btn_upload.data[-1])
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
out_pl
```

```python
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
lbl_pred
```

```python
btn_run = widgets.Button(description='Classify')
btn_run
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
btn_upload = widgets.FileUpload()
```

```python
VBox([widgets.Label('Select your bear!'), btn_upload, btn_run, out_pl, lbl_pred])
```
